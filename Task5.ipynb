#Balancing

```python
df['Class'].value_counts()
```

```python
# Balancing Data using SMOTE
x = df.drop(columns=['Class'])
y = df['Class']
smote = SMOTE()
X, Y = smote.fit_resample(x, y)
data_SMOTE = pd.concat([pd.DataFrame(X), pd.DataFrame(Y)], axis=1)
```

```python
data_SMOTE['Class'].value_counts()
```

```python
sns.countplot(x=data_SMOTE["Class"], palette=["#F9DEDF","#432371"])
plt.title("Data After Balancing", fontsize=15)
plt.show()
```

#Dealing with Outliars

```python
col=['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','V11','V12','V13','V14','V15','V16','V17','V18','V19','V20','V21','V22','V23','V24','V25','V26','V27','V28','Amount','Time']
for a in range(6):
  # Setting Value of outliers to null
  for x in col:
      q25 = data_SMOTE[x].quantile(0.25)
      q75 = data_SMOTE[x].quantile(0.75)
      intr_qr = q75 - q25

      max = q75 + (1.5 * intr_qr)
      min = q25 - (1.5 * intr_qr)

      data_SMOTE.loc[data_SMOTE[x] < min, x] = np.nan
      data_SMOTE.loc[data_SMOTE[x] > max, x] = np.nan
  # Fill nulls with mean
  for x in col:
    data_SMOTE.fillna(statistics.median(data_SMOTE[x]), inplace=True)
print(data_SMOTE.isnull().sum())
```

#Models

```python
# splitting data
x = data_SMOTE.drop(columns='Class', axis=1)
y = data_SMOTE['Class']
X_train, X_test, Y_train, Y_test = train_test_split(x, y, random_state=30, test_size=0.2)
```

```python
print(x.shape, X_test.shape, X_test.shape)
print(y.shape)
```

```python
# LogisticRegression
model = LogisticRegression()
model.fit(X_train, Y_train)
X_test_prediction = model.predict(X_test)

# Evaluation
accuracy = accuracy_score(Y_test, X_test_prediction)
precision = precision_score(Y_test, X_test_prediction)
recall = recall_score(Y_test, X_test_prediction)
f1 = f1_score(Y_test, X_test_prediction)
print(
    "Evaluation metrics on test data of Logistic Regression:\nAccuracy: {}\nF1-Score: {}\nPrecision: {}\nRecall: {}".format(
        accuracy, f1, precision, recall))

cm = confusion_matrix(Y_test, X_test_prediction)
print(cm)
```

```python
# GaussianNaiveBayes
model = GaussianNB()
model.fit(X_train, Y_train)
X_test_prediction = model.predict(X_test)

# Evaluation
accuracy = accuracy_score(Y_test, X_test_prediction)
precision = precision_score(Y_test, X_test_prediction)
recall = recall_score(Y_test, X_test_prediction)
f1 = f1_score(Y_test, X_test_prediction)
print("Evaluation metrics on test data of Naive base:\nAccuracy: {}\nF1-Score: {}\nPrecision: {}\nRecall: {}".format(
    accuracy, f1, precision, recall))

cm = confusion_matrix(Y_test, X_test_prediction)
print(cm)
```

```python
# DecisionTree
DecTree = tree.DecisionTreeClassifier()
DecTree.fit(X_train, Y_train)
y_pred = DecTree.predict(X_test)

# Evaluation
accuracy = accuracy_score(Y_test, y_pred)
precision = precision_score(Y_test, y_pred)
recall = recall_score(Y_test, y_pred)
f1 = f1_score(Y_test, y_pred)
print(
    "Evaluation metrics on test data of Decision Tree:\nAccuracy: {}\nF1-Score: {}\nPrecision: {}\nRecall: {}".format(
        accuracy, f1, precision, recall))

print("Confusion Matrix")
cm = confusion_matrix(Y_test, y_pred)
print(cm)
```
